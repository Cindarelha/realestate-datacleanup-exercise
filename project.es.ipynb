#### Ejercicio 14. Ahora vamos a trabajar con el "cinturón sur" de Madrid. Has un subconjunto del Dataframe original que contenga las siguientes poblaciones (columna level5): "Fuenlabrada","Leganés","Getafe","Alcorcón" (★★☆)

```python
# Definir las poblaciones del "cinturón sur" de Madrid
poblaciones_cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filtrar el DataFrame original usando la columna `level5` y la función `isin`
df_cinturon_sur = df[df['level5'].isin(poblaciones_cinturon_sur)]

# Mostrar el subconjunto filtrado
print(df_cinturon_sur.head())

# Contar el número de filas para cada población
conteo_poblaciones = df_cinturon_sur['level5'].value_counts()
print(conteo_poblaciones)
```

#### Ejercicio 15. Traza un gráfico de barra de la media de los precios y explica lo que observas (debes usar el subconjunto obtenido de la pregunta 14) (★★★)

#### Ejercicio 16. Calcula la media y la varianza dela muestra para las siguientes variables: precio, habitaciones, superficie y baños (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Imprime ambos valores por  cada variable.

```python
# Calcular la media y la varianza para las variables: precio, habitaciones, superficie y baños
variables = ['price', 'rooms', 'surface', 'bathrooms']
estadisticas = {}

for variable in variables:
    media = df_cinturon_sur[variable].mean()
    varianza = df_cinturon_sur[variable].var()
    estadisticas[variable] = {'media': media, 'varianza': varianza}

# Imprimir los valores de media y varianza por cada variable
for variable, stats in estadisticas.items():
    print(f"Variable: {variable}")
    print(f"Media: {stats['media']}")
    print(f"Varianza: {stats['varianza']}\n")
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# Normalizar restando la media y dividiendo por la desviación estandar
df_cinturon_sur['normalized_price'] = df_cinturon_sur.groupby('level5')['price'].transform(lambda x: (x - x.mean()) / x.std())

# Plotear los histogramas normalizados
plt.figure(figsize=(10, 5))

for population in poblaciones_cinturon_sur:
    subset = df_cinturon_sur[df_cinturon_sur['level5'] == population]
    plt.hist(subset['normalized_price'], bins=20, alpha=0.5, label=population)

plt.title('Normalized Price para las poblaciones en el Cinturón Sur de Madrid')
plt.xlabel('Normalized Price')
plt.ylabel('Frecuecia')
plt.legend()
plt.grid(True)
plt.show()
```

# Al normalizarlo nos permite comparar en un orden de escala similar los precios de cada población. Se puede observar la distribución y su parecido con respecto a las diferentes poblaciones.

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in the question 14 (★★☆)

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
# Nueva columna en el df "pps" (price per square meter)
df_cinturon_sur.loc[:, 'pps'] = df_cinturon_sur['price'] / df_cinturon_sur['surface']

# Filtra el df por ciudades "Getafe" y "Alcorcón"
df_getafe = df_cinturon_sur[df_cinturon_sur['level5'] == 'Getafe']
df_alcorcon = df_cinturon_sur[df_cinturon_sur['level5'] == 'Alcorcón']

# Calcula media y varianza 'pps' 
mean_pps_getafe = df_getafe['pps'].mean()
mean_pps_alcorcon = df_alcorcon['pps'].mean()
var_pps_getafe = df_getafe['pps'].var()
var_pps_alcorcon = df_alcorcon['pps'].var()

# Imprimir
print(f"Getafe - Media del precio por metro cuadrado: {mean_pps_getafe:.2f} EUR/m², Varianza: {var_pps_getafe:.2f}")
print(f"Alcorcón - Media del precio por metro cuadrado: {mean_pps_alcorcon:.2f} EUR/m², Varianza: {var_pps_alcorcon:.2f}")
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
# Subplots para cada población
fig, axs = plt.subplots(2, 2, figsize=(14, 10), sharex=True, sharey=True)

# Definir los elementos de población
poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
axs = axs.flatten()

for ax, population in zip(axs, poblaciones):
    subset = df_cinturon_sur[df_cinturon_sur['level5'] == population]
    ax.scatter(subset['surface'], subset['pps'], alpha=0.6, edgecolor='w', color='blue')
    ax.set_title(f'{population}')
    ax.set_xlabel('Superficie (m²)')
    ax.set_ylabel('Precio por m² (EUR/m²)')
    ax.grid(True)

plt.suptitle('Scatter Plots of Price per Square Meter for Different Populations', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
